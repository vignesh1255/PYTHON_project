# 5.Convert datatype of column published at from object to pandas datatime

```python
df['published_at']=df['published_at'].astype('datetime64[ns]')
df.info()
```

# 6.Create a new column as 'published_month' using the column published_at(display the months only)

```python
df['published_month']=pd.DatetimeIndex(df['published_at']).month
df.head()
```

```python
df.tail()
```

# 7.Replace the numbers in the column published_month as names of the months i,e.,  as 'jan', 2 as 'feb' & so on.. 

```python
df['published_month']=df['published_month'].replace({1:'jan',2:'feb',3:'mar',4:'april',5:'may',6:'june',7:'july',8:'aug',9:'sep',10:'oct',11:'nov',12:'dec'})
df.head()
```
